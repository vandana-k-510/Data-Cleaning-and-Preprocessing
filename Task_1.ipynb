```python
df.rename(columns = {'product_id':'Product_Id',
                     'product_name' : 'Product_Name',
                    'category' : 'Category',
                    'discounted_price' :'Discounted_Price',
                    'actual_price' : 'Actual_Price',
                    'discount_percentage' : 'Discount_Percentage',
                    'rating' :'Rating',
                    'rating_count' : 'Rating_Count',
                    'about_product' : 'About_Product',
                    'user_id' : 'User_Id',
                    'user_name':'User_Name',
                    'review_id':'Review_Id',
                    'review_title' : 'Review_Title',
                    'review_content	' : 'Review_Content',
                    'img_link' : 'Img_Link',
                    'product_link' :'Product_Link'}, inplace = True)
```

```python
df.columns
```

```python
# Remove currency symbols and commas, then convert to float

# For Discounted_Price
df['Discounted_Price'] = df['Discounted_Price'].replace('[\₹,]', '', regex=True).astype(float)

# For Actual_Price
df['Actual_Price'] = df['Actual_Price'].replace('[\₹,]', '', regex=True).astype(float)
```

```python
# Remove '%' symbol and convert discount_percentage to float
df['Discount_Percentage'] = df['Discount_Percentage'].replace('%', '', regex=True).astype(float)

# Remove commas and convert rating_count to int
df['Rating_Count'] = df['Rating_Count'].replace(',', '', regex=True).astype(float)

# convert Rating to float
df['Rating'] = df['Rating'].astype(float)
```

```python
# Check which values are not numeric in Rating
df[~df['Rating'].str.replace('.', '', 1).apply(lambda x: x.isdigit())]
```

```python
import numpy as np

# Replace non-numeric ratings with NaN
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
```
